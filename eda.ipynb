```python
import pandas
data = pandas.read_csv("datasets/supermarket_data.csv")
data
```

```python
# use the describe function
data.describe()
```

```python
# we have noticed that the column item price column contains three empty slots.we need to fill these empty slots with a value and in this case we shall fill with the mean of the column

data["itemprice"]= data["itemprice"].fillna(55)

# describe the data again
data.describe()
```

```python
# get the number of entries
data.groupby("payment").size()
```

```python
# combining multiple columns
data.groupby(["city", "payment"]).size()
```

```python
# import the modules

import seaborn as sns

import matplotlib.pyplot as plt
```

```python
# below is a countplot showing the distribution of males to females

sns.countplot(x="gender", data=data ,palette="dark")
plt.title("Gender distribution against total count")
plt.ylabel("Total count of individuals")
```

```python
# Create a count plot of product
sns.countplot(x="product", data=data ,palette="autumn")
plt.xticks(rotation= 45)
plt.title("Product distribution")
# The product category that appears the most
```

```python
# below is a count plot that has twoo columns in it
sns.countplot(x="gender", hue="city", data=data ,palette="spring")
```

```python
# Below is a pie plot 
data["payment"].value_counts().plot(kind="pie", autopct="%1.1f%%")
```
